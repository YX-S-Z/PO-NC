```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

with open('cifar10-cubic-0/info.pkl', 'rb') as f:
    cifar_10_cubic = pickle.load(f)

with open('cifar10-exp-0/info.pkl', 'rb') as f:
    cifar_10_exp = pickle.load(f)

with open('cifar10-linear-0/info.pkl', 'rb') as f:
    cifar_10_linear = pickle.load(f)

with open('cifar10-log-0/info.pkl', 'rb') as f:
    cifar_10_log = pickle.load(f)

with open('cifar10-quadratic-0/info.pkl', 'rb') as f:
    cifar_10_quad = pickle.load(f)

with open('cifar10-sqrt-0/info.pkl', 'rb') as f:
    cifar_10_sqrt = pickle.load(f)

with open('cifar10-uniform-0/info.pkl', 'rb') as f:
    cifar_10_uni = pickle.load(f)

with open('mnist-cubic-0/info.pkl', 'rb') as f:
    mnist_10_cubic = pickle.load(f)

with open('mnist-exp-0/info.pkl', 'rb') as f:
    cifar_10_exp = pickle.load(f)

with open('cifar10-linear-0/info.pkl', 'rb') as f:
    cifar_10_linear = pickle.load(f)

with open('cifar10-log-0/info.pkl', 'rb') as f:
    cifar_10_log = pickle.load(f)

with open('cifar10-quadratic-0/info.pkl', 'rb') as f:
    cifar_10_quad = pickle.load(f)

with open('cifar10-sqrt-0/info.pkl', 'rb') as f:
    cifar_10_sqrt = pickle.load(f)

with open('cifar10-uniform-0/info.pkl', 'rb') as f:
    cifar_10_uni = pickle.load(f)
```

```python
fig, axs = plt.subplots(2, 4, figsize=(16, 8))
name_list = ['collapse_metric', 'ETF_metric', 'WH_relation_metric', 'Wh_b_relation_metric', 'train_acc1', 'train_acc5', 'test_acc1', 'test_acc5']

for j in range(2):
    for i in range(4):
        # Get the x y for plotting
        x = range(len(cifar_10_uni[name_list[4*j+i]]))
        y1 = cifar_10_cubic[name_list[4*j+i]]
        y2 = cifar_10_exp[name_list[4*j+i]]
        y3 = cifar_10_linear[name_list[4*j+i]]
        y4 = cifar_10_log[name_list[4*j+i]]
        y5 = cifar_10_quad[name_list[4*j+i]]
        y6 = cifar_10_sqrt[name_list[4*j+i]]
        y7 = cifar_10_uni[name_list[4*j+i]]
    
        # Plot the subfigures
        axs[j,i].plot(x, y1, label='cubic')
        axs[j,i].plot(x, y2, label='exp')
        axs[j,i].plot(x, y3, label='linear')
        axs[j,i].plot(x, y4, label='log')
        axs[j,i].plot(x, y5, label='quad')
        axs[j,i].plot(x, y6, label='sqrt')
        axs[j,i].plot(x, y7, label='uni')
        
        # Set legend and title
        if j==0:
            axs[j,i].set_title('CIFAR10: NC_{}, {}'.format(i,name_list[i]))
        else:
            axs[j,i].set_title('{}'.format(name_list[4*j+i]))   
        axs[j,i].set_xlabel('Epoch')
        axs[j,i].set_ylabel('Value')
        axs[j,i].legend()
        axs[j,i].grid()
# Adjust the spacing between subplots
fig.tight_layout()

# Show the plot
plt.show()
```
