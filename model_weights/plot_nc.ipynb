```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

with open('35125-Mon-Jun-26-16-54/info.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
# Generate some random data for the curves
x = range(len(data['collapse_metric']))
y1 = data['collapse_metric']
y2 = data['ETF_metric']
y3 = data['WH_relation_metric']
y4 = data['Wh_b_relation_metric']

# Create a new figure
plt.figure()

# Plot the curves with different colors
plt.plot(x, y1, label='NC_1')
plt.plot(x, y2, label='NC_2')
plt.plot(x, y3, label='NC_3')
# plt.plot(x, y4, label='NC_4')

# Set the title and labels
plt.title('Four Curves')
plt.xlabel('Epoch')
plt.ylabel('Value')

# Add a legend
plt.legend()

# Add grid
plt.grid(True)


# Display the plot
plt.show()
```
